#### 1. Find the download website and version
Windows download website: https://www.mysql.com/downloads/



- Choose: MySQL Installer for Windows

- Download: The newest version or the version you like, click download

#### 2. Install it
- Choose 'custom', keep 'Next'
- Add all 6 items to the right
- Till you see picture below, almost done
  


- Creat your own password and remember it
- Apply 'configuration', when finished (all turn to green lights), click 'cancel' to exit

#### 3. Build MySQL server
- Windows 'command prompt' and 'run as administrator'

```python
'net start mysql80' # start the server to connect MySQL
'net stop mysql80' # stop the server to connect MySQL
```

#### 4. Start MySQL
- Start to use MySQL method1: use MySQL command and enter the password

- Start to use MySQL method2: Windows 'command prompt' to enter and set 'PATH' (search for how to set PATH with this method to start)

```python
mysql [-h 127.0.01] [-P 3306] -u root -p # 'P' means 'port'; 'u' means 'user' ; 'p' means 'password'
```
